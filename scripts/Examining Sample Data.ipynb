```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rc("figure", figsize=(10,10))
os.environ["PROJ_LIB"] = "C:\\Utilities\\Python\\Anaconda\\Library\\share"; #fixr
```

```python
data_directory = '/data/notebooks/jupyterlab-biascorrlab/data/'
```

```python
data = np.load(f"{data_directory}single_site_jan_precip.npy")  # This is 38 years of daily data for January
print(data.shape,data)
```

```python
# Plotting the Probability Distribution as a Histogram

bins = np.arange(0, 5, 0.1)
p = len(data[data != 0]) / len(data)
print(f'Probability of Snowfall = {round(1-p,2)}') 
weights = np.full(len(data[data != 0]), p)

plt.hist(
    data[data != 0],
    bins=bins,
    histtype="step",
    stacked=True,
    fill=False,
    weights=weights,
)

plt.xlabel("Daily Snowfall", fontsize=12)
plt.ylabel("Probability Density", fontsize=12)
plt.show()
```
