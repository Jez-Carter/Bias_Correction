```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.set_prop_cycle('color',plt.cm.Spectral(np.linspace(0,1,25)))

xs = np.linspace(0.01,1,100)
kappa=1

for phi in np.arange(0,5,0.2):
    ys = np.exp(-(phi*xs)**kappa)
    plt.plot(xs,ys,label=f'Phi={round(phi,2)}')
    
plt.legend()
plt.title(f'Varying Phi, Kappa={kappa}')
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.set_prop_cycle('color',plt.cm.Spectral(np.linspace(0,1,25)))

xs = np.linspace(0.01,2,100)
phi=1.0

for kappa in np.arange(0,5,0.2):
    ys = np.exp(-(phi*xs)**kappa)
    plt.plot(xs,ys,label=f'Kappa={round(kappa,2)}')

ax.set_ylim([0, 1])
plt.legend()
plt.title(f'Varying Kappa, Phi={phi}')
```

```python
fig, ax = plt.subplots(figsize=(20,15))
ax.set_prop_cycle('color',plt.cm.Spectral(np.linspace(0,1,25)))

xs = np.linspace(0.01,4,100)
for phi in [0.5,1]:
    for kappa in np.arange(0,5,0.2):
        ys = np.exp(-(phi*xs)**kappa)
        plt.plot(xs,ys,label=f'Phi={round(phi,2)},Kappa={round(kappa,2)}')

ax.set_ylim([0, 1])
plt.legend()
plt.title(f'Varying Kappa and Phi')
```
