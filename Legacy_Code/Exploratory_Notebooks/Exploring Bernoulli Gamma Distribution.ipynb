```python
import numpy as np
from scipy.stats import gamma
import matplotlib.pyplot as plt
from src.model_fitting_functions import BernoulliGamma
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.set_prop_cycle('color',plt.cm.Spectral(np.linspace(0,1,13)))
xs = np.linspace(0.01,0.5,100)
p,alpha,beta = 0.7,0.5,2
for alpha_adjustment in np.arange(-0.4,0.8,0.1):
    alpha_new = alpha + alpha_adjustment
    dist = BernoulliGamma((p,alpha_new,beta))
    ys = np.exp(dist.log_prob(xs.reshape(1,-1)))
    plt.plot(xs,ys,label=f'Alpha={round(alpha_new,2)}')
plt.legend()
plt.title('Varying Alpha')
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.set_prop_cycle('color',plt.cm.Spectral(np.linspace(0,1,11)))
xs = np.linspace(0.1,3,100)
p,alpha,beta = 0.7,0.5,2
for beta_adjustment in np.arange(-1.2,1,0.2):
    beta_new = beta + beta_adjustment
    dist = BernoulliGamma((p,alpha,beta_new))
    ys = np.exp(dist.log_prob(xs.reshape(1,-1)))
    plt.plot(xs,ys,label=f'Beta={round(beta_new,2)}')
plt.legend()
plt.title('Varying Beta')
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.set_prop_cycle('color',plt.cm.Spectral(np.linspace(0,1,9)))
xs = np.linspace(0.1,3,100)
p,alpha,beta = 0.7,0.5,2
for beta_adjustment in np.arange(-1,1.01,1):
    ax.set_prop_cycle('color',plt.cm.Spectral(np.linspace(0,1,4)))
    for alpha_adjustment in np.arange(-0.4,0.41,0.4):
        alpha_new = alpha + alpha_adjustment
        beta_new = beta + beta_adjustment
        dist = BernoulliGamma((p,alpha_new,beta_new))
        ys = np.exp(dist.log_prob(xs.reshape(1,-1)))
        plt.plot(xs,ys,label=f'Alpha={round(alpha_new,2)},Beta={round(beta_new,2)}')
plt.legend()
plt.title('Varying Alpha and Beta')
```
